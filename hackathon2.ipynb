```python
#Function for getting the top ten similar users for a given user
def get_similarity(user_id,neighbors=5):
    similarity = pd.DataFrame(user_user_matrix[user_id])
    similarity["user_id"] = list(user_user_matrix)
    similarity = similarity.sort_values(by=user_id,ascending=False)
    similarity = similarity[similarity[user_id]<1.0]
    return similarity[['user_id',user_id]].head(neighbors)
```

```python
def get_challenges_(user):
    return train.loc[train.user_id==user,'challenge'].tolist()
```

```python
def get_item_similarity_(challenge1,challenge2):
    if (challenge1 != challenge2):
        return corr_matrix.loc[challenge1,challenge2]
    else:
        print('Cannot compare to self.')
```

```python
def get_recommendations_(user_id,n=3):
    users = get_similarity(user_id)
    user_1 = get_challenges_(users.user_id.iloc[0])
    user_2 = get_challenges_(users.user_id.iloc[1])
    user_3 = get_challenges_(users.user_id.iloc[2])
    user_4 = get_challenges_(users.user_id.iloc[3])
    user_5 = get_challenges_(users.user_id.iloc[4])
    target_user = get_challenges_(user_id)
    user_1 = [x for x in user_1 if x not in target_user]
    user_2 = [x for x in user_1 if x not in target_user]
    user_3 = [x for x in user_1 if x not in target_user]
    user_4 = [x for x in user_1 if x not in target_user]
    user_5 = [x for x in user_1 if x not in target_user]
    return user_1[:n]
```
